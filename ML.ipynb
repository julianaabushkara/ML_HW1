```python
import pandas as pd

\ Correct path to the dataset
df = pd.read_csv('dataset_migdal/migdal_1_sift_dataset.csv')

# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Check for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Check data types
print("\nData types:\n", df.dtypes)

print("\nNumber of rows:",df.head())


print("\ndescribe:",df.describe())

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot distribution of 'Response'
sns.histplot(df['Response'], bins=50, kde=True,color="teal")
plt.title('Response Distribution')
plt.show()

# Plot distribution of 'Scale'
sns.histplot(df['Scale'], bins=50, kde=True)
plt.title('Scale Distribution')
plt.show()

# Plot one of the SIFT features (e.g., Feature_1)
sns.histplot(df['Feature_1'], bins=50, kde=True,color='purple')
plt.title('Distribution of SIFT Feature 1')
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import seaborn as sns



# ---------- DISTRIBUTION VISUALIZATIONS ----------
plt.figure(figsize=(16, 4))

plt.subplot(1, 3, 1)
sns.histplot(df['Response'], kde=True, bins=30, color='teal')
plt.title("Response Distribution")
plt.xlabel("Response")

plt.subplot(1, 3, 2)
sns.histplot(df['Scale'], kde=True, bins=30, color='coral')
plt.title("Scale Distribution")
plt.xlabel("Scale")

plt.subplot(1, 3, 3)
sns.histplot(df['Feature_1'], kde=True, bins=30, color='purple')
plt.title("SIFT Feature 1 Distribution")
plt.xlabel("Feature_1")

plt.tight_layout()
plt.show()

# ---------- KEYPOINT VISUALIZATION ----------


img = cv2.imread('dataset_migdal/migdal_1.JPG')  # Use correct case and extension

if img is None:
    raise FileNotFoundError("❌ Image not found. Please check the filename and path.")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Plot keypoints
plt.figure(figsize=(10, 10))
plt.imshow(img_rgb)
plt.quiver(df['X'], df['Y'],
           np.cos(np.radians(df['Angle'])),
           np.sin(np.radians(df['Angle'])),
           angles='xy', scale_units='xy', scale=0.5,
           color='lime', width=0.002)
plt.title('Keypoints on migdal_1.JPG')
plt.gca().invert_yaxis()
plt.axis('off')
plt.show()

```
